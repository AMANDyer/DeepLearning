```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
df.dtypes
```

```python
x=df.drop(columns=['Attrition'], axis=1)
y=df['Attrition']
```

```python
x.info
```

```python
categorical_features=['BusinessTravel',
                      'Department',
                      'EducationField',
                      'Gender',
                      'JobRole',
                      'MaritalStatus',
                      'Over18',
                      'OverTime']
```

```python
x_encoded=pd.get_dummies(x, columns=categorical_features, drop_first=True )
```

```python
x_encoded=x_encoded.astype(int)
x_encoded
```

```python
lb=LabelEncoder()
```

```python
y_encoded=lb.fit_transform(y)
```

```python
x_train, x_test, y_train, y_test=train_test_split(x_encoded, y_encoded, test_size=0.2, random_state=42)
```

```python
x_train.head()
```

```python
sc=StandardScaler()
```

```python
x_train_scaled=sc.fit_transform(x_train)
x_test_scaled=sc.transform(x_test)
```

Now we will create architecture of our model


```python
model=tf.keras.models.Sequential([
    tf.keras.layers.Dense(units=124, activation='relu', input_dim=x_train_scaled.shape[1]),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(units=62, activation='relu'),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(units=1, activation='sigmoid')
])
```

```python
# Compile the Model
```

```python
loss_fn=tf.keras.losses.BinaryFocalCrossentropy()
metrics=['accuracy']
```

```python
learning_rate=0.001
momentum=0.9
optimizer=tf.keras.optimizers.SGD(learning_rate=learning_rate,
                                  momentum=momentum, nesterov=True)
```

```python
model.compile(optimizer=optimizer, loss=loss_fn, metrics=metrics)
```

```python
# Train the model
```

```python
model.fit(x_train_scaled,
          y_train,
          epochs=50,
          batch_size=8,
          validation_split=0.1)
```

```python
loss, accuracy=model.evaluate(x_test_scaled, y_test)
```

```python
print(f'loss: {loss}, accuracy: { accuracy}')
```
