```
df["quality_label"] = df["quality"].apply(lambda x: 1 if x >= 7 else 0)
```

```
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
```

```
sc=StandardScaler()
lb=LabelEncoder()
x=df.drop(columns=["quality","quality_label"])
y=df["quality_label"]
```

```
from scipy.sparse import random
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```
df.isnull().sum()
```

```
x_train_scaled=sc.fit_transform(x_train)
x_test_scaled=sc.transform(x_test)
y_train_scaled=lb.fit_transform(y_train)
```

```
x_train_scaled
```

```
import tensorflow as tsf
from sklearn.utils.class_weight import compute_class_weight
```

```
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train),
    y=y_train
)
class_weight_dict = {0: class_weights[0], 1: class_weights[1]}
```

```
class_weight_dict
```

```
model=tsf.keras.models.Sequential([
    tsf.keras.layers.Dense(units=124,activation='relu',input_dim=x_train_scaled.shape[1]),
    tsf.keras.layers.Dropout(0.3),
    tsf.keras.layers.Dense(64,activation='relu'),
    tsf.keras.layers.Dropout(0.3),
    tsf.keras.layers.Dense(1,activation='sigmoid')
])
```

```
model.compile(optimizer='adam',loss=tsf.keras.losses.binary_crossentropy,metrics=['accuracy'])
```

```
model.fit(x_train_scaled,y_train_scaled,epochs=10,batch_size=30,class_weight=class_weight_dict)
```

```
loss,accuracy=model.evaluate(x_test_scaled,y_test)
```
