```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()

def stem_words(text):
  return " ".join([ps.stem(word) for word in text.split()])
```

```python
sample="walk walks walking walked"
stem_words(sample)
```

```python
text="probably my alltime favourite movie story of selflessness and dedication"
stem_words(text)
```

```python
from nltk.stem import WordNetLemmatizer

import nltk
nltk.download('wordnet')
```
