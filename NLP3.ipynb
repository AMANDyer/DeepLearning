```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame({'text':['people watch cricket','player was player','people write comment','cricket play comment'],'output':[1,1,0,0]})
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
```

```python
bow=cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
```

```python
cv.transform(["cricket watch and write comment of cricket comment"]).toarray()
```

```python
#cv=CountVectorizer(ngram_range=(1,2))
v=CountVectorizer(ngram_range=(1,2))
```

```python
bow=v.fit_transform(df['text'])
```

```python
df['text']
```

```python
print(v.vocabulary_)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf=TfidfVectorizer()
tfidf.fit_transform(df['text']).toarray()
```

```python
print(tfidf.idf_)
print(tfidf.get_feature_names_out())
```
